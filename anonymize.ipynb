### Anonimizacja danych

```python
from anonympy.pandas import dfAnonymizer # https://github.com/ArtLabss/open-data-anonymizer
# przykłady https://colab.research.google.com/drive/1wg4g4xWTSLvThYHYLKDIKSJEC4ChQHaM?usp=sharing#scrollTo=VhHUGjftoA-U

orginal_df = df.copy() # kopia oryginalnych danych

ano_df = dfAnonymizer(df)
ano_df
```

```python
print('Categorical Columns: ', ano_df.categorical_columns)
print('Datetime Columns: ', ano_df.datetime_columns)
print('Numeric Columns: ', ano_df.numeric_columns)
```

```python
# anonimiacja danych kategorycznych przez tokenizację 
# Tokenization works by removing the valuable data from your environment and replacing it with these tokens
# Unlike encrypted data, tokenized data is undecipherable and irreversible.
# https://www.tokenex.com/resource-center/what-is-tokenization
# tj. każda wartość, która się będzie powtarząć będzie miała taki sam token, ale nie jest to operacja odwracalna
ano_df.categorical_tokenization(ano_df.categorical_columns, inplace=True)
```

```python
# anonimiacja danych daty przez pertubację / dodanie "szumu"
# może mieć znaczy wpływ na działanie modeli ML ponieważ znacznie zmienia zależności
# wpisy, które różniły się minutą, po pertubacj mogą różnić się dniami
# stąd należy rozważyć usunięcie daty całkowicie
# ew. potrakotować jako wartość numeryczną i użyć anomizacji dla danych numerycznych
ano_df.datetime_noise(ano_df.datetime_columns, frequency=('DAY', 'HOUR', 'minutes'), inplace=True)
```

```python
# anonimiacja danych numerycznych przez maskownie PCA 
# https://medium.com/lizuna/beacon-the-use-of-principal-components-analysis-to-mask-sensitive-data-in-machine-learning-7904b01445d0
# Another slightly more complex example of data masking can be performed using Principal Component Analysis, or PCA. 
# PCA was developed as a means of reducing the number of variables necessary to take into account when analyzing data. 
# It accomplishes this by creating new “principal components” from linear transformations of the original variables.
# Each observation would have a value for that principal component, which explains a certain amount of variation in the data, 
# and is thus useful to the machine learning algorithm. If those numeric values are sent without an explanation of how they were calculated, 
# the algorithm could still be trained on these values and make more accurate predictions, 
# but malicious actors would not know how to interpret those masked values.
ano_df.numeric_masking(ano_df.numeric_columns, inplace=True)
 
```
